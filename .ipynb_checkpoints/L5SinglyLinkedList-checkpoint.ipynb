# Linked Lists

## Singly Linked List
- A collection of nodes that collectively form a linear sequence
- Each node stores the reference to the next node of the list



In a Linked List the first node is called the **head** and the last node is called the **tail**. Let's discuss the pros and cons of Linked Lists:

## Pros

* Linked Lists have constant-time insertions and deletions in any position, in comparison, arrays require O(n) time to do the same thing.

* Linked lists can continue to expand without having to specify their size ahead of time (remember on Array sizing form the Array Sequence section)

## Cons

* To access an element in a linked list, you need to take O(n) time to go from the head of the list to the nth element. In contrast, arrays have constant time operations to access elements in an array.

```python
class Node():
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class LinkedList():
    def __init__(self):
        self.head = None
        self.tail = None
        
    def append(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            self.tail = self.head
            self.length = 1
        else:
            self.tail.next = new_node
            self.tail = new_node 
            self.length += 1
            
    def prepend(self,data):
        new_node = Node(data)
        new_node.next = self.head 
        self.head = new_node
        
        self.length += 1
        
        
    def insert(self,index,data):
        new_node = Node(data)
        i = 0
        temp = self.head
        if index>=self.length:
            self.append(data)
            return 
        
        if index ==0:
            self.prepend(data)
            return
        
        while i<self.length:
            if i == index-1:
                temp.next , new_node.next = new_node , temp.next
                self.length+=1
                break
            
            temp = temp.next
            i+=1
    
    
    def remove(self,index):
        temp = self.head
        i=0
        
        if index>=self.length:
            print("Entered wrong index")
            
        if index == 0:
            self.head = self.head.next
            self.length -= 1   
            return       
        
        while i<self.length:
            if i == index-1:
                temp.next = temp.next.next
                self.length-=1
                break
            
            i+=1
            temp = temp.next
            
    def reverse(self):
        prev = None
        self.tail = self.head 
        while self.head != None:
            temp = self.head
            self.head = self.head.next
            temp.next = prev
            prev = temp  
        
        self.head = temp
        
```
